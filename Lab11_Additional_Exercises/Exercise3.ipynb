```python
import sqlite3
import os
import shutil

conn = sqlite3.connect("profile_management.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS profiles (
    user_id INTEGER PRIMARY KEY AUTOINCREMENT,
    display_name TEXT NOT NULL,
    image_filename TEXT NOT NULL
)
""")

conn.commit()

if not os.path.exists("uploads"):
    os.makedirs("uploads")

# USER INPUT
display_name = input("Enter display name: ")

image_path = input("Enter image file path: ")

allowed_extensions = ('.jpg', '.jpeg', '.png')

if image_path.lower().endswith(allowed_extensions):

    filename = os.path.basename(image_path)

    destination = os.path.join("uploads", filename)

    try:

        shutil.copy(image_path, destination)

        cursor.execute("""
        INSERT INTO profiles(display_name, image_filename)
        VALUES (?, ?)
        """, (display_name, filename))

        conn.commit()

        print("\nProfile updated successfully!")

    except FileNotFoundError:

        print("\nImage file not found!")

else:

    print("\nInvalid image format!")

print("\nSAVED PROFILES")

cursor.execute("""
SELECT * FROM profiles
""")

for row in cursor.fetchall():
    print(row)

cursor.close()
conn.close()

print("\nDatabase connection closed.")
```
