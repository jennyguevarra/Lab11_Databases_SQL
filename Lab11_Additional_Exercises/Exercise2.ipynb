```python
import sqlite3
from datetime import date

conn = sqlite3.connect("attendance.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS attendance (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    student_name TEXT NOT NULL,
    attendance_date TEXT NOT NULL,
    status TEXT NOT NULL
)
""")

conn.commit()

student_name = input("Enter student name: ")

attendance_status = input("Enter attendance status (Present/Absent): ")

today = date.today()

cursor.execute("""
INSERT INTO attendance(student_name, attendance_date, status)
VALUES (?, ?, ?)
""", (student_name, str(today), attendance_status))

conn.commit()

print("\nAttendance recorded successfully!")

print("\nATTENDANCE RECORDS")

cursor.execute("""
SELECT * FROM attendance
""")

for row in cursor.fetchall():
    print(row)

search_name = input("\nEnter student name to search: ")

cursor.execute("""
SELECT * FROM attendance
WHERE student_name LIKE ?
""", ('%' + search_name + '%',))

print("\nSEARCH RESULT")

for row in cursor.fetchall():
    print(row)

cursor.close()
conn.close()

print("\nDatabase connection closed.")
```
